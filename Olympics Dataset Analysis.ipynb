```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athletes = pd.read_csv('/Users/santul/Downloads/archive-2/athlete_events.csv')
regions = pd.read_csv('/Users/santul/Downloads/archive-2/country_definitions.csv')
```

```python
regions.head()
```

```python
athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
athletes_df.rename(columns={'region':'Region','notes':'Notes'}, inplace = True)
```

```python
# Top 10 countries participating 

top_10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
# PLOTTING THE TOP 10 COUNTRIES PARTICIPATED IN OLYMPICS

plt.figure(figsize =(12,6))
# plt.xticks(rotation = 20)
plt.title('Overall participation by country')
sns.barplot(x= top_10_countries.index, y= top_10_countries, palette ='Set2')
```

```python
# AGE DISTRIBUTION OF THE PARTICIPANTS

plt.figure(figsize = (12,6))
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'orange', edgecolor = 'white');
```

```python
# Winter olympic sports
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# Summer olympic sports
summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# Male and Female Participants

gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pieplot for male and female participants

plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle = 150, shadow = True)
```

```python
# Total medals won

athletes_df.Medal.value_counts()
```

```python
Female_participants = athletes_df[(athletes_df.Sex == 'F')&(athletes_df.Season == 'Summer')][['Sex' , 'Year']]
Female_participants = Female_participants.groupby('Year').count().reset_index()
Female_participants.tail()
```

```python
WomenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data = WomenOlympics, palette="Spectral")
plt.title('Women Participation')
```

```python
part = WomenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
# Gold medal athletes

goldmedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldmedals.head()
```

```python
# Athletes which have gold medals beyond age 60

goldmedals['ID'][goldmedals['Age']>60].count()
```

```python
# what sporting event does they have won

sporting_event = goldmedals['Sport'][goldmedals['Age']>60]
sporting_event
```

```python
# plot for sporting event

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for Athletes over 60 years')
```

```python
# Gold medals for each country

goldmedals.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalGoldMedals = goldmedals.Region.value_counts().reset_index(name='Medal').head()
g = sns.catplot(x = "index", y = "Medal", data=totalGoldMedals,height=5, kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country')
```

```python
 # Analyzing 2016 Rio Olympics
    
max_year = athletes_df.Year.max()
max_year

team_list = athletes_df[(athletes_df.Year == max_year)& (athletes_df.Medal == 'Gold')].Team
team_list.value_counts().head(10)
```

```python
sns.barplot(x=team_list.value_counts().head(20), y=team_list.value_counts().head(20).index)

plt.xlabel(None);
plt.xlabel('Countrywise Medals for the year 2016');
```
